```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import denoise_functions.Denoise_Functions as df
import denoise_functions.Filters as ft
from denoise_functions.Filters import overall_analysis
```

## Define Variables

```python
temp = np.linspace(-10, 10, 20001)
func = df.Denoise_Functions(temp)
f1_plot = func.f1_plot()
f2_plot = func.f2_plot()
f3_plot = func.f3_plot()
f4_plot = func.f4_plot()
f5_plot = func.f5_plot()
f1 = func.f1()
f2 = func.f2()
f3 = func.f3()
f4 = func.f4()
f5 = func.f5()
```

## Plot Functions

```python
plt.figure(num=None, figsize=(16, 10), dpi=80, facecolor='w', edgecolor='k')
plt.subplot(2, 3, 1)
plt.plot(temp, f1_plot, label='function 1', linewidth=1)
plt.title('Function #1')
plt.legend()
plt.subplot(2, 3, 2)
plt.plot(temp, f2_plot, label='function 2', linewidth=1)
plt.title('Function #2')
plt.legend()
plt.subplot(2, 3, 3)
plt.plot(temp, f3_plot, label='function 3', linewidth=1)
plt.title('Function #3')
plt.legend()
plt.subplot(2, 3, 4)
plt.plot(temp, f4_plot, label='function 4', linewidth=1)
plt.title('Function #4')
plt.legend()
plt.subplot(2, 3, 5)
plt.plot(temp, f5_plot, label='function 5', linewidth=1)
plt.title('Function #5')
plt.legend()
```
